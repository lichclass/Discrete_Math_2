$\textbf{V. Bipartite Graphs}$

This is a bipartite graph because its nodes can be divided into two disjoint sets such that every edge connects a node from one set to a node in the other set, without any edges connecting nodes within the same set.

```python
import matplotlib.pyplot as plt
import networkx as nx

mySeed = 50

node_names = ['A', 'B', 'C', 'D', 'E']
name_map = {i: name for i, name in enumerate(node_names)}
G = nx.bipartite.gnmk_random_graph(2, 3, 6)
G = nx.relabel_nodes(G, name_map)

# Separate nodes into two sets
set1 = {node for node, node_data in G.nodes(data=True) if node_data['bipartite'] == 0}
set2 = set(G) - set1

# Draw bipartite graph
pos = nx.spring_layout(G, seed=mySeed)
nx.draw(G, pos, nodelist=set1, node_color='skyblue', node_size=1500)
nx.draw(G, pos, nodelist=set2, node_color='lightgreen', node_size=1500)
nx.draw_networkx_labels(G, pos=pos)

plt.title("Example of a Bipartite Graph")
plt.show()
```

$\textbf{VIII. Handshake Theorem}$

It states that in any finite undirected graph, the sum of the degrees of all vertices is twice the number of edges.
Equation:

$deg(v)=2e$, such that deg(v) = degree of a vertex, and e = total number of edges

Example from the Previous Graph:

$Edges(e) = 5$

$TotalDegrees(t) = 10$

Thus,

$t=2(5)=10$

$\textbf{XII. Euler's Circuit}$

This graph is an Euler's Circuit because it satisfies the following conditions:

>if every vertex in the graph has an even degree.

>The graph is connected

```python
import matplotlib.pyplot as plt
import networkx as nx

node_names = ['A', 'B', 'C', 'D', 'E', 'F']
edge_list = [("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), ("E", "F"), ("F", "A"), ("A", "C"), ("C", "E"), ("A", "E")]
G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edge_list)

euler_c = list(nx.eulerian_circuit(G))
pos = nx.spring_layout(G, seed=35)

nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=100)
for u, v in euler_c:
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    dx = x1 - x0
    dy = y1 - y0
    mid_x = (x0 + x1) / 2
    mid_y = (y0 + y1) / 2
    arrow_length_factor = 0.2
    plt.arrow(mid_x, mid_y, dx * arrow_length_factor, dy * arrow_length_factor, head_width=0.02, head_length=0.06, fc='red', ec='red')

plt.title("Graph with Eulerian Circuit and Arrows")
plt.show()
```

$\textbf{XIII. Euler's Trail}$

This graph is an Euler's Trail because it satisfies the following conditions:

>The starting Node(C) and the ending Node(F) have odd degrees, while the rest have even degrees

>The graph is connected

```python
import matplotlib.pyplot as plt
import networkx as nx

node_names = ['A', 'B', 'C', 'D', 'E', 'F']
edge_list = [('A', 'B'), ('A', 'F'), ('C', 'B'), ('C', 'D'), ('C', 'F'), ('E', 'D'), ('E', 'F')]
G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edge_list)

euler_path = list(nx.eulerian_path(G))
pos = nx.planar_layout(G)

nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=100)
for u, v in euler_path:
        x0, y0 = pos[u]
        x1, y1 = pos[v]
        dx = x1 - x0
        dy = y1 - y0
        mid_x = (x0 + x1) / 2
        mid_y = (y0 + y1) / 2
        arrow_length_factor = 0.2
        plt.arrow(mid_x, mid_y, dx * arrow_length_factor, dy * arrow_length_factor, head_width=0.02, head_length=0.06, fc='red', ec='red')

plt.title("Graph with Eulerian Trail and Arrows")
plt.show()
```

$\textbf{XIV. Hamiltonian Circuits}$

The graph is a Hamiltonian Circuit because it satisfies the following conditions:

>The path contains every subgraph of the original graph

>The path is connected

>The path has the same number of edges as vertices

>Every vertex of the path has degree 2

```python
import matplotlib.pyplot as plt
import networkx as nx

def hamilton(G):
    F = [(G,[list(G.nodes())[0]])]
    n = G.number_of_nodes()
    while F:
        graph,path = F.pop()
        confs = []
        neighbors = (node for node in graph.neighbors(path[-1])
                     if node != path[-1]) #exclude self loops
        for neighbor in neighbors:
            conf_p = path[:]
            conf_p.append(neighbor)
            conf_g = nx.Graph(graph)
            conf_g.remove_node(path[-1])
            confs.append((conf_g,conf_p))
        for g,p in confs:
            if len(p)==n:
                return p
            else:
                F.append((g,p))
    return None

node_names = ['A', 'B', 'C', 'D']
edge_list = [("A", "B"), ("B", "C"), ("C", "D"), ("D", "A"), ("A", "C"), ("B", "D")]

G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edge_list)

# Find a Hamiltonian cycle

hamiltonian_c = list(hamilton(G))

# Visualize the graph
# Visualize the graph
pos = nx.spring_layout(G, seed=35)
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=100)

# Draw edges from Hamiltonian circuit
for i in range(len(hamiltonian_c) - 1):
    u, v = hamiltonian_c[i], hamiltonian_c[i + 1]
    nx.draw_networkx_edges(G, pos=pos, edgelist=[(u, v)], arrows=True, arrowsize=20, edge_color='red')

u, v = hamiltonian_c[-1], hamiltonian_c[0]
nx.draw_networkx_edges(G, pos=pos, edgelist=[(u, v)], arrows=True, arrowsize=20, edge_color='red')

plt.title("Graph with Hamiltonian Circuit")
plt.show()
```

$\textbf{XV. Travelling Salesman Problem}$

```python
import matplotlib.pyplot as plt
import networkx as nx

node_names = ["A", "B", "C", "D"]
edges = [('A', 'B', {'weight': 30}),
         ('A', 'C', {'weight': 50}),
         ('A', 'D', {'weight': 40}),
         ('B', 'C', {'weight': 30}),
         ('B', 'D', {'weight': 35}),
         ('C', 'D', {'weight': 25})]

routes = [
    ['A', 'B', 'C', 'D', 'A'],
    ['A', 'B', 'D', 'C', 'A'],
    ['A', 'C', 'B', 'D', 'A'],
    ['A', 'C', 'D', 'B', 'A'],
    ['A', 'D', 'B', 'C', 'A']
]

fig, axs = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    G = nx.Graph()
    G.add_nodes_from(node_names)
    G.add_edges_from(edges)
    pos = nx.planar_layout(G)

    nx.draw_networkx_edge_labels(
        G, pos,
        edge_labels={('A', 'B'): '30',
                     ('A', 'C'): '50',
                     ('A', 'D'): '40',
                     ('B', 'C'): '30',
                     ('B', 'D'): '35',
                     ('C', 'D'): '25'},
        font_color='red', ax=axs[i]
    )

    nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=1500, ax=axs[i])

    # Highlighting the specified route
    for j in range(len(routes[i]) - 1):
        nx.draw_networkx_edges(
            G, pos, edgelist=[(routes[i][j], routes[i][j + 1])],
            edge_color='red', width=2, ax=axs[i]
        )

    # Calculating the weight of the route
    weight = sum(G[routes[i][j]][routes[i][j + 1]]['weight'] for j in range(len(routes[i]) - 1))
    axs[i].set_title('Route {}'.format(i + 1))
    axs[i].text(0.5, -0.15, 'Weight: {}'.format(weight), horizontalalignment='center', verticalalignment='center',
                transform=axs[i].transAxes)

# Adjust layout
plt.tight_layout()
plt.gcf().canvas.manager.set_window_title("TRAVELLING SALESMAN PROBLEM")
plt.show()
```

$\textbf{XV. Travelling Salesman Problem - Example}$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the nodes and edges
nodes = ['A', 'B', 'C', 'D', 'E']

edge_labels = {
    ('A', 'B'): 10,
    ('A', 'C'): 20,
    ('A', 'D'): 30,
    ('A', 'E'): 40,
    ('B', 'C'): 50,
    ('B', 'D'): 60,
    ('B', 'E'): 70,
    ('C', 'D'): 80,
    ('C', 'E'): 90,
    ('D', 'E'): 100,
    ('E', 'D'): 5
}

# Convert edge labels to the desired format
formatted_edge_labels = {edge: "{} feet".format(weight) for edge, weight in edge_labels.items()}

# Define Hamilton circuits
hamilton_circuits = [
    ['A', 'B', 'C', 'D', 'E', 'A'],
    ['A', 'B', 'D', 'E', 'C', 'A'],
    ['A', 'B', 'E', 'D', 'C', 'A'],
    ['A', 'E', 'D', 'C', 'B', 'A'],
    ['A', 'E', 'D', 'B', 'C', 'A']
]

# Plotting
fig, axs = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    G = nx.Graph()
    G.add_nodes_from(nodes)

    # Add edges with formatted labels
    for edge, label in formatted_edge_labels.items():
        G.add_edge(edge[0], edge[1], weight=edge_labels[edge])

    pos = nx.circular_layout(G)

    nx.draw_networkx_edge_labels(
        G, pos,
        edge_labels=formatted_edge_labels,
        font_color='red', ax=axs[i]
    )

    nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=1500, ax=axs[i])

    # Highlighting the specified Hamilton circuit
    for j in range(len(hamilton_circuits[i]) - 1):
        nx.draw_networkx_edges(
            G, pos, edgelist=[(hamilton_circuits[i][j], hamilton_circuits[i][j + 1])],
            edge_color='red', width=2, ax=axs[i]
        )

    # Calculating and displaying the weight of the route
    weight = sum(G[hamilton_circuits[i][j]][hamilton_circuits[i][j + 1]]['weight'] for j in range(len(hamilton_circuits[i]) - 1))
    axs[i].set_title('Route {}'.format(i + 1))
    axs[i].text(0.5, -0.15, 'Weight: {}'.format(weight), horizontalalignment='center', verticalalignment='center',
                transform=axs[i].transAxes)

# Adjust layout
plt.tight_layout()
plt.show()
```
